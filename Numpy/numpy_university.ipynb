```python
over_30 = len(arr[arr>30])/len(arr)*100
over_30
```

Створіть матрицю розміром 20 × 4 із нормального розподілу із середнім μ= 10 та
стандартним відхиленням σ = 2.5.

```python
matrix  = np.random.normal(loc=10, scale=2.5, size=(20,4))
matrix
```

Знайдіть суму елементів першого рядка

```python
np.sum(matrix[0])
```

Знайдіть максимальний елемент останнього стовпця

```python
np.max(matrix[:,-1])
```

Знайдіть середнє значення у кожному рядку

```python
np.mean(matrix, axis=1)
```

Створіть нову матрицю того ж розміру, елементи якої обчислюються за правилом: bij = aij − min, де aij - елементи вихідної матриці, min - її мінімальний елемент.

```python
min_value = matrix.min()
new_matrix = matrix - min_value
new_matrix
```

Знайдіть рядки, третій елемент яких більший за 5

```python
new_matrix[new_matrix[:,2]>5]
```

```python
np.where(new_matrix[:, 2] > 5)
```

Знайдіть стовпці, максимальний елемент яких більший за 12

```python
np.where(np.max(new_matrix,axis=0)>12)
```

```python
np.max(new_matrix,axis=0)>12
```
